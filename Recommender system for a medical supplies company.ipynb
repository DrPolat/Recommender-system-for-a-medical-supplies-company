```python
def popular(column_name,ranking="total"):
    counts=df[column_name].value_counts().head(10).index.tolist()
    grouped=df[df[column_name].isin(counts)].groupby(column_name).agg(
        Quantity=("Order_Items.qty",np.sum),
        Price=("Order_Items.price",np.mean))
       
    grouped["Total_Amount"]=grouped["Quantity"]*grouped["Price"]
    
    if ranking == "sum":
        return grouped.sort_values("Quantity",ascending=False).head(10)
    
    else:
        return grouped.sort_values("Total_Amount",ascending=False).head(10)
    
    
```

## Cosine similarity for product description

```python
user_item_matrix=df.pivot_table(index="Orders.customer_id",
                                columns="Order_Items.product_name",
                               values="Order_Items.qty").fillna(0.0)
```

```python
user_item_matrix
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity

#ürün-özellik matrixi
vect=CountVectorizer()
item_features=vect.fit_transform(df["Order_Items.product_name"])

#özellikleri ağırlıklandırma
tfidf=TfidfTransformer()
item_features_weighted=tfidf.fit_transform(item_features)

#ürünlerin arasındaki benzerliği hesaplama
item_similarity=cosine_similarity(item_features_weighted)
```

```python
#örnek olarak 10.ürünün benzer ürünlerini görüntüleyin
similar_items=item_similarity[10].argsort()[::-1]
print(similar_items)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# ürün benzerlik matrisini hesaplayın
item_similarity_matrix = cosine_similarity(user_item_matrix.T.fillna(0))


# 10. ürünün benzer ürünlerini bulun
item_id = 10
similar_item_ids = np.argsort(item_similarity_matrix[item_id-1])[::-1][1:]
similar_items = user_item_matrix.columns[similar_item_ids]

```
